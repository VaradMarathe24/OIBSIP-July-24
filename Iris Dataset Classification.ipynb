Import the required libraries into the code.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

Loading the data

```python
flower=pd.read_csv("Iris.csv")
print(flower)
print(flower.shape)
```

```python
print(flower.describe())
print(flower.isna().sum())
print(flower.describe())
print(flower.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(flower.drop('Species',axis=1),flower['Species'],test_size=0.2,random_state=0)
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
x_new=np.array([[5,2.9,1,0.2,2]])

```

```python
prediction=knn.predict(x_new)
print("Prediction:{}".format(prediction))
```

```python
#check accuracy of the model
from sklearn.svm import SVC
model1=SVC()
model1.fit(x_train,y_train)
y_pred=model1.predict(x_test)
from sklearn.metrics import accuracy_score
print('Accuracy=',accuracy_score(y_test,y_pred))
```

```python
# Create a bar chart showing the number of flowers for each species
species_counts = flower['Species'].value_counts()
species_counts.plot(kind='bar')
plt.title("Number of Flowers belonging to the respective species")
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
sepal_lengths = flower['SepalLengthCm'].value_counts()
sepal_widths = flower['SepalWidthCm'].value_counts()

# Create a bar chart for sepal lengths
plt.figure(figsize=(10, 5))
sepal_lengths.plot(kind='bar')
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')
plt.show()

# Create a bar chart for sepal widths
plt.figure(figsize=(10, 5))
sepal_widths.plot(kind='bar')
plt.title('Sepal Width Distribution')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Count')
plt.show()
```
